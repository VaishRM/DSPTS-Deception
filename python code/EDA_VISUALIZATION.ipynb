```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the combined data from CSV
data_all = pd.read_csv('physio_data_all.csv')
```

```python
# Create a new figure for boxplots
plt.figure(figsize=(15, 10))
```

```python
# Create boxplots for each parameter
for i, param in enumerate(['HR', 'HRV', 'SCL', 'SCR']):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='deception_block', y=param, data=data_all)
    plt.title('Boxplot of ' + param)
```

```python
plt.tight_layout()
plt.show()
```

```python
# Create a new figure for histograms
plt.figure(figsize=(15, 10))
```

```python
# Create histograms for each parameter
for i, param in enumerate(['HR', 'HRV', 'SCL', 'SCR']):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=data_all, x=param, hue="deception_block", element="step", stat="density", common_norm=False)
    plt.title('Histogram of ' + param)
```

```python
plt.tight_layout()
plt.show()
```
